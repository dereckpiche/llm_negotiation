```python
import numpy as np
import matplotlib.pyplot as plt

betas = np.floor(np.random.beta(0.01, 0.01, 1000) * 15 + 1)
#print(betas)

plt.hist(betas, bins=np.arange(0, 16, 1))
plt.show()


```
